```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('/content/Leena.png')
img_copy = np.copy(img)
```

```python
img_copy = cv2.cvtColor(img_copy,cv2.COLOR_BGR2RGB)
plt.imshow(img_copy)
```

```python
gray = cv2.cvtColor(img_copy,cv2.COLOR_RGB2GRAY)
plt.imshow(gray,cmap='gray')
```

CUSTOM FILTER

```python
kernel_enhanced_edge = np.array([[-1,0,1],
                  [-2,0,2],
                  [-1,0,1]])
```

```python
filtered_img_enh_edge = cv2.filter2D(gray,-1,kernel_enhanced_edge)
```

```python
plt.imshow(filtered_img_enh_edge)
```

EDGE DETECTION

```python
kernel_Edge = np.array([[0,1,0],
                  [1,-2,1],
                  [0,1,0]])
```

```python
filtered_img_edge = cv2.filter2D(gray,-1,kernel_Edge)
```

```python
plt.imshow(filtered_img_edge)
```
